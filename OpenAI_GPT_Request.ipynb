```python
import openai
import os
```

```python
openai.organization = "YOUR ORGANIZATION"
openai.api_key = "YOUR API Key"
```

```python
def respond_letter(text):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
                {"role": "system", "content": "You are the recipient of the following message. Write a message that responds to the sender. Use \"<NAME>\" as the placeholder for any names."},
                {"role": "user", "content": text},

        ],
        max_tokens=300,
    )
    response_text = response['choices'][0]['message']['content']
    return response_text
```

```python
import os
import regex as re


msg_dir = "."
msg_lst = []
for file in os.listdir(msg_dir):
    print(file)
    with open(file, 'r') as f:
        msg_lst.extend(f.read().split("***\n"))

msg_lst = [msg for msg in msg_lst if len(msg) > 0]
```

Create Inputs and Responses

```python
import pandas as pd
import numpy as np
```

```python
inputs = msg_lst[:]
outputs = [respond_letter(msg) for msg in inputs]
data = [[inputs[i], outputs[i]] for i in range(len(inputs))]
```

```python
df = pd.DataFrame(data, columns=['Inputs', 'Outputs'])
```
